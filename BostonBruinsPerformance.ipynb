We wanted to see how the Boston Bruins performed between 1990 and 2011 using this data set. We wanted to analyze if

```python
Bruins_df= df[df['Team Name'] == "Boston Bruins"]
Leafs_df = df[df['Team Name'] == "Toronto Maple Leafs"]
Bruins_df
```

Below is my P2 assignment where I wanted to see how the Boston Bruins win percentage was related to their goal
differential. I wanted to know if the higher their goal differential was the higher the win percentage. I also wanted to 
find out if more losses was correlated to more goals against.

```python
x = Bruins_df['+ / -'].values.reshape(-1, 1)
y = Bruins_df['Win %'].values
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
w = Bruins_df['Goals Against (GA)'].values.reshape(-1, 1)
z = Bruins_df['Losses'].values
model2 = LinearRegression()
model2.fit(w, z)
z_pred = model2.predict(w)
```

```python
df = pd.read_csv("NHL_Scraped_data.csv")
```

```python
Bruins_df = df[df['Team Name'] == "Boston Bruins"]
```

```python
fig, (a, b) = plt.subplots(1, 2, figsize=(16, 6))
a.scatter(Bruins_df['+ / -'], Bruins_df['Win %'], color='blue', label='Data Points')
a.plot(Bruins_df['+ / -'], y_pred, color='red', label='Linear Regression Line')
a.set_xlabel('Goal Differential')
a.set_ylabel('Win Percentage')
a.set_title('Boston Bruins: Goal Differential vs. Win Percentage')
a.legend()
a.grid(True)
b.scatter(w, z, color='blue', label='Data Points')
b.plot(w, z_pred, color='red', label='Linear Regression Line')
b.set_xlabel('Goals Against (GA)')
b.set_ylabel('Losses')
b.set_title('Boston Bruins: Losses vs. Goals Against')
b.legend()
b.grid(True)
```

Goal differential and win percentage exhibit a positive correlation, where higher goal differentials are associated with higher win percentages. Similarly, losses and goals against show a strong positive correlation, with more goals allowed leading to more losses. Maintaining consistency over such a long period is challenging, but the data reveals significant insights into the Boston Bruins historical success and game strategy over these two decades.

Below is my P3 assignment where I wanted to compare the Bruins win percentage over time to the Toronto Maple Leafs 
win percentage since the Maple Leafs are one of the Bruins main rivals. The visualization below shows how the Boston 
Bruins and the Toronto Maple Leafs performance from 1990 to 2000 was relatively similar but from 2005 to 2011 the Bruins
on average well outperformed the Maple Leafs including winning a Stanley Cup in 2011.

```python
plt.figure(figsize=(10, 6))
plt.plot(Bruins_df['Year'], Bruins_df['Win %'], marker='o', linestyle='-', color='b', label='Bruins Win %')
plt.plot(Leafs_df['Year'], Leafs_df['Win %'], marker='o', linestyle='-', color='r', label='Leafs Win %')
plt.xlabel('Year')
plt.ylabel('Win Percentage')
plt.title('Boston Bruins Compared to Toronto Maple Leafs Performance Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

We implemented a time series graph to visualize performance data from 1990 to 2011. Our goal was to compare the Bruins
win percentage to that of one of their main rivals, the Toronto Maple Leafs. Despite fluctuations, most of the assumptions
we can draw from this graph focus on the late 2000s, indicating a noticeable shift in momentum. The Maple Leafs display
greater variance, with significant dips in certain years, while the Bruins exhibit a smoother upward trend over time,
though they also experience fluctuations in win percentage. These inconsistencies highlight major changes in team 
dynamics throughout the two decades.
